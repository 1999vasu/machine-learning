```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
monkey_data = np.random.multivariate_normal([1.0,2.0],[[1.5,0.5],[0.5,1]],1000)
chimp_data = np.random.multivariate_normal([4.0,4.0],[[1,0],[0,1.8]],1000)

print(monkey_data.shape)
print(chimp_data.shape)
plt.scatter(monkey_data[:,0],monkey_data[:,1],label='Monkey')
plt.scatter(chimp_data[:,0],chimp_data[:,1],alpha=.5,label='Chimpanzee')
query_pt = np.array([6,4])
plt.scatter(query_pt[0],query_pt[1],c='k',label="query point")
```

```python
data = np.zeros((2000,3))
data[:1000,:-1] = monkey_data[:,:]
data[1000:,:-1] = chimp_data[:,:]
data[1000:,-1] = 1
print(data.shape)
```

```python
x_train = data[:,:-1]
y_train = data[:,-1]

print(x_train.shape,y_train.shape)
    
```

```python
def dist(x1,x2):
    return np.sqrt(np.sum((x2-x1)**2))


def knn(x_train,y_train,q_point,k=11):
    dist_val = []
    
    m = x_train.shape[0]
    for ix in range(m):
        tempdist = dist(q_point,x_train[ix,:])
        dist_val.append((tempdist,y_train[ix]))
        
    dist_val = sorted(dist_val)
    dist_val = dist_val[:k]
#     print(dist_val)
    
    y = np.array(dist_val)
#     print(y.shape)
    te = np.unique(y[:,1],return_counts=True)
#     print(te)
    idx = te[1].argmax()
#     print(idx)
    prediction = te[0][idx]
    return prediction


knn(x_train,y_train,query_pt)
    
```

```python
def accuracy(xtrain,ytrain,xtest,ytest):
    acc = 0
    for ix in range(xtest.shape[0]):
        knn_pred = knn(xtrain,ytrain,xtest[ix])
#         print(ix,knn_pred,ytest[ix])
        if(knn_pred == ytest[ix]):
            acc+=1
    return acc/xtest.shape[0]
```

```python
print(accuracy(x_train,y_train,x_train,y_train))
```
