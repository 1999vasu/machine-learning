```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
```

```python
epoch = 20
lr = 0.01

num_input = 784
num_classes = 10

layer_1_nodes = 128
layer_2_nodes = 64

display_step = 5
mnist = input_data.read_data_sets( 'data', one_hot=True )
```

```python
with tf.variable_scope('input'):
    X = tf.placeholder(dtype = tf.float32, shape = [None,num_input])
    
with tf.variable_scope('layer1'):
    W1 = tf.Variable(tf.random_normal(shape =[num_input,layer_1_nodes]),name ='weight_1')
    B1 = tf.Variable(tf.random_normal(shape = [layer_1_nodes] ),name = 'bias_1')
    layer1out = tf.matmul(X,W1) + B1
    layer1out = tf.nn.relu(layer1out)

with tf.variable_scope('layer2'):
    W2 = tf.Variable(tf.random_normal(shape =[layer_1_nodes,layer_2_nodes]),name ='weight_2')
    B2 = tf.Variable(tf.random_normal(shape = [layer_2_nodes] ),name = 'bias_2')
    layer2out = tf.matmul(layer1out,W2) + B2
    layer2out = tf.nn.relu(layer2out)

with tf.variable_scope('output'):
    W_out = tf.Variable( tf.random_normal( shape= [layer_2_nodes, num_classes] ), name = 'weights_out' )
    B_out = tf.Variable( tf.zeros( shape= [num_classes] ), name = 'biases_out' )
    output = tf.matmul( layer2out, W_out ) + B_out
    
with tf.variable_scope('cost'):
    Y = tf.placeholder( shape=[None, num_classes], dtype=tf.float32 )
    cost = tf.nn.softmax_cross_entropy_with_logits_v2( logits=output, labels=Y )
    cost = tf.reduce_mean( cost )

with tf.variable_scope('train'):
    opt = tf.train.AdamOptimizer(learning_rate=lr).minimize(cost)

train_writer = tf.summary.FileWriter('./training')
train_writer.add_graph( tf.get_default_graph() )

with tf.variable_scope('logging'):
    summary = tf.summary.scalar( 'cost', cost )

saver = tf.train.Saver()    
```

```python
with tf.Session() as sess:
    sess.run( tf.global_variables_initializer() )

#     last_ckpt = tf.train.latest_checkpoint('./weights')
#     saver.restore( sess, last_ckpt )

    for epoch in range(epoch):

        for _ in range(100):

            batch = mnist.train.next_batch(32)

            sess.run( opt, feed_dict= {
                X: batch[0],
                Y: batch[1]
            } )

        batch = mnist.train.next_batch(256)
        training_loss, train_summary = sess.run( [cost, summary] , feed_dict={
            X: batch[0],
            Y: batch[1]
        } )

        train_writer.add_summary( train_summary, epoch )


        test_loss = 0.0
        for _ in range(50):
            batch = mnist.test.next_batch(32)
            test_loss += sess.run( cost, feed_dict={
                X: batch[0],
                Y: batch[1]
            } )

        test_loss = test_loss/50

        if ( epoch + 1 ) % display_step == 0:
            print ( 'Training Loss is ' + str(training_loss) + ' Testing loss is ' + str(test_loss) + '\n')
            saver.save(sess, './weights/epoch_' + str(epoch + 1) + '.ckpt')
            print (" Model Saved \n ")
            print (" ----------- \n")
```
