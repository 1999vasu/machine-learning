```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

```python
x,y = make_regression(n_features=2,n_samples=200,bias=4,noise=1)
```

```python
print(x.shape,y.shape)
```

```python
X = np.ones((x.shape[0],x.shape[1]+1))
print(X.shape)
```

```python
X[:,1:]=x
```

```python
def hypothesis(theta,x): #calculates hypothesis for individual data
    return np.dot(theta,x)


def error(theta,x,y): #calculates error for all dataset
    err=0;
    m = x.shape[0];
    for ix in range(m):
        hyp = hypothesis(theta,x[ix,:])
        err += (y[ix]-hyp)**2
    return err/(2*x.shape[0])

def gradient(theta,x,y):  #calculates gradient
    grad = np.zeros(theta.shape[0])
    m = x.shape[0];
    for ix in range(m):
        hyp = hypothesis(theta,x[ix,:])
        grad += (y[ix]-hyp)*x[ix,:]
    return grad/m

def gradientdescent(x,y,n=0.001):  #gradient descend algorithm
    theta = np.zeros(x.shape[1])
    err = [1]
    while(True):
        temp = error(theta,x,y)
        err.append(temp)
        print(temp)
        theta = theta + n*gradient(theta,x,y)
        if(err[-2]-err[-1]<0.01 and err[-2]-err[-1]>=0):
            break
    err = err[1:]
    return theta,err
    
```

```python
theta,er = gradientdescent(X,y)
er = np.array(er)
# print(er)
th = np.arange(er.shape[0])
# print(th)
```
