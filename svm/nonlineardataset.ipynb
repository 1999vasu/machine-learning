```python
import numpy as np
from sklearn.datasets import make_circles
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x,y = make_circles(n_samples= 500, noise = 0.05)
```

```python
print(x.shape,y.shape)
```

```python
plt.scatter(x[:,0],x[:,1],c = y)
plt.show()
```

```python
def xconvert(x):
    x_ = np.zeros((x.shape[0],x.shape[1]+1))
    x_[:,:-1]= x
    x_[:,-1] = x[:,0]**2+x[:,1]**2
    return x_
```

```python
print(x[:3,:])
x_ = xconvert(x)
print(x_[:3,:])
```

```python
def plot3d(x,Y):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111,projection='3d')
    ax.scatter(x[:,0],x[:,1],x[:,2],zdir = 'z',c=Y,s=20,depthshade=True)
    plt.show()
```

```python
plot3d(x_,y)
```

# NORMAL LOGISTIC CLASSIFIER

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

```python
lr = LogisticRegression()
```

```python
print('accuracy is {}'.format(cross_val_score(lr,x,y,cv=5).mean()))
```

# Logistic classifier on higher dimension

```python
print('accuracy is {}'.format(cross_val_score(lr,x_,y,cv=5).mean()))
```
