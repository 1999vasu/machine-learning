```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
dataX = pd.read_csv('./weightedX.csv')
dataY = pd.read_csv('./weightedY.csv')
```

```python
X = dataX.values
Y = dataY.values
```

```python
print(X.shape,Y.shape)
```

```python
plt.scatter(X[:,0],Y[:,0],color = 'orange')
```

```python
#Normalizing data
u = X.mean()
sd = X.std()
X = (X-u)/sd
```

```python
plt.scatter(X[:,0],Y[:,0],color = 'orange')
```

```python
def getW(qp,x,y,tau):
    m = x.shape[0]
    
    W = np.mat(np.eye(m))
    for ix in range(m):
        xi = x[ix]
        temp =np.dot((xi-qp),(xi-qp).T)
#         print(temp.shape)
        W[ix,ix] = np.exp(temp[0,0]/(-2*tau*tau))
    
    return W
```

```python
def prediction(qp,X,Y,tau):
    ones = np.ones((X.shape[0],1))
    X = np.hstack((X,ones))
    X = np.mat(X)
    Y = np.mat(Y)
    qp = np.mat(np.hstack((qp,[1])))
    
    W = getW(qp,X,Y,tau)
    
    first = (X.T*W*X)
    second = (X.T*W*Y)
    
    theta = np.linalg.pinv(first)*second
    pred = qp*theta
    return theta,pred[0,0]
    
```

```python
theta,pred = prediction(X[50],X,Y,2)
```

```python
print(pred,Y[50])
```

```python
#visualization
```

```python
def visualize(x,y,tau):
    
    xtest = np.linspace(-2,2,20)
    ytest = []
    
    for ix in xtest:
        _,pred = prediction(ix,x,y,tau)
        ytest.append(pred)
    
    ytest = np.array(ytest)
    
    plt.scatter(x,y,c='b')
    print(xtest.shape,ytest.shape)
    plt.scatter(xtest,ytest,c='r')
    plt.show()
```

```python
visualize(X,Y,2.0)
visualize(X,Y,1.5)
visualize(X,Y,1.0)
visualize(X,Y,0.5)
visualize(X,Y,0.3)

```
