```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
monkey_data = np.random.multivariate_normal([1.0,2.0],[[1.0,0.5],[0.5,1.0]],1000)
print(monkey_data.shape)
```

```python
cheempanzee_data = np.random.multivariate_normal([0.0,3.0],[[1.0,0.8],[0.8,1.0]],1000)
print(cheempanzee_data.shape)
```

```python
plt.scatter(monkey_data[:,0],monkey_data[:,1],c='r',label="monkey")
plt.scatter(cheempanzee_data[:,0],cheempanzee_data[:,1],c='g',label="cheempanzee")
plt.legend()
plt.show()
```
